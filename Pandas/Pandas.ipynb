````python
d1 = {
    'A' : [10,20,30,40],
    'B' : [3,4,5,6]
}
d2 = {
    'A' : [30,40,50,60],
    'C' : [3,4,5,6]
}
````

````python
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2,index=[0,1,2,5])
````

````python
df1
````

````python
df2
````

````python
# df3 = pd.concat([df1,df2],ignore_index=True)
# df3 = pd.concat([df1,df2],axis = 1)
# df3 = pd.concat([df1,df2],join='inner')
df3 = pd.concat([df1,df2],join='inner',axis = 1)
df3
````

````python
mdf3 = pd.merge(df1,df2,on='A',how='outer')
mdf3
````

````python
mdf3.count()
````
