```python
res = requests.get('https://webscraper.io/test-sites/tables')
```

```python
res
```

```python
print(res.text)
```

```python
soup = bs4.BeautifulSoup(res.text)
```

```python
print(soup.prettify())
```

```python
table_tags = soup.find_all('table')
```

```python
tb = table_tags[0]
```

```python
tb
```

```python
# tb.find('thead')
table_data = []
h = True
for tb in table_tags:
    row = []
    if h:
        for th in tb.thead.find_all('th'):
            row.append(th.text)
        
        table_data.append(row)
        h = False
    
    for tr in tb.tbody.find_all('tr'):
        row = []
        for td in tr.find_all('td'):
            row.append(td.text)
    
        table_data.append(row)
table_data
```

```python
table_data
```

```python
import pandas as pd
tdata = pd.DataFrame(table_data[1:],columns=table_data[0])
```

```python
tdata
```

```python
with open('Data.csv','w') as file:
    for row in table_data:
        file.write(','.join(row)+'\n')
```
