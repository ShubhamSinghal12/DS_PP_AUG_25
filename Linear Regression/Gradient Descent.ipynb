```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fn(x):
    y = 4*x**2 - 2*x + 3
    return y
def derF(x):
    return 8*x-2
```

```python
x = np.linspace(-3.5,4,100)
y = fn(x)
plt.plot(x,y)
plt.show()
```

```python
x = 3
alpha = 0.001
xs = [x]
```

```python
for i in range(1000):
    # print(x)
    x = x - alpha*derF(x)
    xs.append(x)
```

```python
x = np.linspace(-3.5,4,100)
y = fn(x)
plt.plot(x,y)
plt.plot(xs,fn(np.array(xs)),'r')
plt.xlim(-4,5)
plt.ylim(0,50)
plt.show()
```

```python
def hypo(w,w0,x):
    # return w[0]*x[:,0] + w[1]*x[:,1] + w0
    return w[0]*x[:,0] + w0
```

```python
# X = np.array([[1,2],[2,1],[3,2],[4,3]])
# y = np.array([4,3,8,13])

X = np.array([[1],[2],[3],[4],[5]])
y = np.array([3,5,7,9,11])
```

```python
# w = [1,1]
w = [1]
w0 = 1
alpha = 0.1
```

```python
hypo(w,w0,X)
```

```python
for e in range(100):
    ypred = hypo(w,w0,X)
    w[0] = w[0] - alpha*np.mean((ypred-y)*X[:,0])
    # w[1] = w[1] - alpha*np.mean((ypred-y)*X[:,1])
    w0 = w0 - alpha*np.mean((ypred-y))

    # print(f'w1 = {w[0]}, w2 = {w[1]}, w0 = {w0}')
    print(f'w1 = {w[0]}, w0 = {w0}')
```
